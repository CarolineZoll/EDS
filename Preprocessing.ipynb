```python
import pandas as pd
import numpy as np
import bibliothek as bib
```

```python
df_ue = pd.read_json('Prepro/ue_data_051222.json')
```

```python
df2 = pd.read_json("Prepro/cell_data.txt")
```

```python
cells=df2['cell']
cell_list=np.array([])
for i in cells:
    cell_list=np.append(cell_list,i['pci'])
cell_list=cell_list.astype(int)
```

```python
data=pd.DataFrame()
```

```python
x=np.array([])
for i in df_ue['id']:
    dic=df_ue.loc[i]['cell_info']
    powers={}
    powers.update({'x':0})
    powers.update({'y':0})
    for i in dic:
        powers.update({i['pci']:i['dl_rx_power']})
    x=np.append(x,powers)
        
```

```python
data = pd.DataFrame.from_records(x)
```

```python
noise=np.power(10,-103/10)
sectors=cell_list
cluster=sectors

df2=pd.DataFrame()
df2=bib.get_int(data,sectors,cluster)
```

```python
df=df2.copy()

SINR_1_list=np.array([])
SINR_2_list=np.array([])
c_cell=np.array([])
s_cell=np.array([])
for i in np.arange(0, len(df)):
    p_r=df.loc[i,:]
    p_r=p_r[2:2+len(sectors)] #goes through file line by line and selects the line
    p_r=np.power(10, p_r/10) # calculate the linear values for capacity calculations
    p_r=p_r.sort_values(ascending=False) # sort values by their order
    a_r=np.sqrt(p_r) #calculate the amplitude
    index=p_r.index.values #Extract all index values 
    index=index.astype(np.int) #PCIen as integer
    a_r=np.array(a_r)
    p_r=np.array(p_r)
    inter=df.loc[i,'inter']
    pci_coord= index[1]
    
    SINR_1= p_r[0]/(np.sum(p_r[1:len(p_r)])+inter+noise) #one bs sending -> rest interference
    SINR_2= np.power(np.sum(a_r[0:2]),2)/(np.sum(p_r[2:len(p_r)])+inter+noise)
    print(SINR_1)
    print(SINR_2)
    SINR_1_list=np.append(SINR_1_list,10*np.log10(SINR_1)) #appends the calculated elements
    SINR_2_list=np.append(SINR_2_list,10*np.log10(SINR_2))
    s_cell=np.append(s_cell,index[0])
    c_cell=np.append(c_cell,index[1])
```

```python
s_cell=s_cell.astype(int)
c_cell=c_cell.astype(int)
```

```python
df['JT_1 SINR [dB]']=SINR_1_list
df['JT_2 SINR [dB]']=SINR_2_list
df['PCI Serving']=s_cell
df['PCI Coord']=c_cell
```
