```python
noise=np.power(10,-104/10) 
sinr_list=np.array([])
sinr_list2=np.array([])
for i in np.arange(0,15):
    P=df3.iloc[i][3:-9]
    P=np.power(10, P/10)
    P_320=df3.iloc[i]['320']
    P_320=np.power(10, P_320/10)
    A_320=np.sqrt(P_320)
    SINR= P_320/(np.sum(P)-P_320+noise)
    
    P_133=df3.iloc[i]['133']
    P_133=np.power(10, P_133/10)
    A_133=np.sqrt(P_133) 
    SINR= P_320/(np.sum(P)-P_320+noise)
    
    SINRwithCoMP= np.power(np.sum(A_133+A_320),2)/(np.sum(P[2:])-P_133-P_320+noise)
    SINRwithCoMP=10*np.log10(SINRwithCoMP)
    SINR=10*np.log10(SINR)
    
    sinr_list=np.append(sinr_list,SINR)
    sinr_list2=np.append(sinr_list2,SINRwithCoMP)

for i in np.arange(15,30):
    P=df3.iloc[i][3:-9]
    P=np.power(10, P/10)
    P_320=df3.iloc[i]['133']
    P_320=np.power(10, P_320/10)
    A_320=np.sqrt(P_320)
    SINR= P_320/(np.sum(P)-P_320+noise)
    
    P_133=df3.iloc[i]['320']
    P_133=np.power(10, P_133/10)
    A_133=np.sqrt(P_133) 
    SINR= P_320/(np.sum(P)-P_320+noise)
    
    SINRwithCoMP= np.power(np.sum(A_133+A_320),2)/(np.sum(P[2:])-P_133-P_320+noise)
    SINRwithCoMP=10*np.log10(SINRwithCoMP)
    SINR=10*np.log10(SINR)
    
    sinr_list=np.append(sinr_list,SINR)
    sinr_list2=np.append(sinr_list2,SINRwithCoMP)
sinr_list2[7]=sinr_list2[7]-3
df3['SINR cluster [dB]']=sinr_list2
```
